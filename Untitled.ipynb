```python
import pandas as pd
import numpy as np
import pinecone
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv("anime.csv")
df = df.dropna(subset=['genre'])
df['rating'] = df['rating'].fillna(df['rating'].mean())
df.dropna(inplace=True)

df['name'] = df['name'].str.strip().str.lower()
df['genre'] = df['genre'].apply(lambda x: x.replace(" ", "") if isinstance(x, str) else x)
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def store_embeddings():
    anime_ids = df['anime_id'].astype(str).tolist()
    genres = df['genre'].tolist()
    
    batch_size = 32  # Adjust based on system performance
    for i in range(0, len(genres), batch_size):
        batch_genres = genres[i:i+batch_size]
        batch_ids = anime_ids[i:i+batch_size]
        batch_embeddings = model.encode(batch_genres).tolist()
        
        index.upsert(vectors=list(zip(batch_ids, batch_embeddings)))

        print(f"Uploaded {i + len(batch_genres)} / {len(genres)} records")

store_embeddings()
```

```python
def find_similar_anime_by_name(anime_name, top_k=5):
    anime_name = anime_name.strip().lower()
    anime_row = df[df['name'] == anime_name]
    
    if anime_row.empty:
        return ["‚ùå Anime not found in the dataset."]

    genre = anime_row.iloc[0]['genre']
    query_embedding = model.encode(genre).tolist()

    results = index.query(vector=query_embedding, top_k=top_k, include_metadata=True)

    # Convert IDs to names
    anime_ids = [match['id'] for match in results['matches']]
    anime_names = df[df['anime_id'].astype(str).isin(anime_ids)]['name'].tolist()

    # Format the output list nicely
    formatted_output = [f"üî• {i+1}. {name.capitalize()}" for i, name in enumerate(anime_names)]
    
    return formatted_output


```

```python
# Example usage
anime_name = "naruto"
similar_anime = find_similar_anime_by_name(anime_name)
print("\n".join(similar_anime)) 
```
